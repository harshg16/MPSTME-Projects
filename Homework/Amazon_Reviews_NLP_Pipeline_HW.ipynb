```python
import pandas as pd
import re
import nltk
import spacy
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_json(r'C:\Users\harsh\Documents\MPSTME\Sem VI\NLP\Office_Products_5.json', lines=True)
df.head()
```

```python
REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')

def clean_text(text):
    text = text.lower()
    text = REPLACE_BY_SPACE_RE.sub(' ', text)
    text = BAD_SYMBOLS_RE.sub('', text)
    return text

df['reviewText'] = df['reviewText'].apply(clean_text)
df['summary'] = df['summary'].apply(clean_text)
```

```python
def spacey(l):
    d = {}
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(l)
    token = []
    lemma = []
    pos = []
    dep = []
    noun = []

    for x in doc:
        token.append(x.text)
        lemma.append(x.lemma_)
        pos.append(x.pos_)
        dep.append(x.dep_)

    for noun1 in doc.noun_chunks:
        noun.append(noun1.text)

    d['Tokens'] = token
    d['Lemmatizer'] = lemma
    d['Pos Tag'] = pos
    d['Dependancy Parser'] = dep
    d['Nouns'] = noun

    return d

df['Spacy']=df.iloc[0:50]['reviewText'].apply(spacey)
df.head()
```
